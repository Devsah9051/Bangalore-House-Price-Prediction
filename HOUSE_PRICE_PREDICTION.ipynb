```python
#droping the values which not make impact the value of house:
df.drop(columns=['area_type','availability','society'],inplace=True)
```

```python
#filling the values where value is null :
#first in location :
df['location'] = df["location"].fillna("Sarjapur  Road ")
#handel thenull value in size:
df['size'] = df['size'].fillna("2 BHK")
#handel now the bathroom :
df['bath'] = df['bath'].fillna(df['bath'].median())
```

```python
#handle the bhk problem :
df["bhk"]=df['size'].str.split().str.get(0).astype(int)
```

```python
#checking the range of total_sqft":
df["total_sqft"].unique()
```

```python
#handle the range of the total_sqft:
def convert_range(x):
  temp=x.split('-')
  if len(temp)==2:
    return (float(temp[0])+float(temp[1]))/2 #getting the mean here of values:
  try:
    return float(x)
  except:
    return None

df["total_sqft"]=df["total_sqft"].apply(convert_range)
```

# Price Per Square feet

```python
#making new columns here:
df["price_per_sqft"]=df['price']*100000/df['total_sqft']
df['price_per_sqft']
```

```python
df.head()
df['bath'].describe()#max=16
df[['bath','price']].corr()
```

```python
#reduce the number of location :
df['location']=df['location'].apply(lambda x:x.strip())#remove the space from front and back:
location_count=df['location'].value_counts()
location_count
```

```python
location_count_lessthan10=location_count[location_count<=10]
location_count_lessthan10
df['location']=df['location'].apply(lambda x:'Others' if x in location_count_lessthan10 else x)
```

```python
df['location'].value_counts()
```

# Outlier handle :

```python
#handle the outlier in total_sqft;
(df['total_sqft']/df['bhk']).describe()
```

```python
#now putting all the value which only have total_sqft>=300 are actual flat:
df=df[(df['total_sqft']/df['bhk'])>=300]
df.describe()
```

```python
df.shape
```

```python
#here making the subdf for making the price_per_sqft into range of mean+std to mean-stdL:
def remove_outlier_sqft(df):
  df_output=pd.DataFrame()
  for key,subdf in df.groupby('location'):
    m=np.mean(subdf.price_per_sqft)
    st=np.std(subdf.price_per_sqft)
    gen_new_df=subdf[(subdf.price_per_sqft>=(m-st))&(subdf.price_per_sqft<=(m+st))]
    df_output=pd.concat([df_output,gen_new_df],ignore_index=True)
  return df_output


df=remove_outlier_sqft(df)
df.describe()
```

```python
#handleing the outlier in the bhk:
def bhk_outlier_remove(df):
  exclude_indices=np.array([])
  for location,location_df in df.groupby("location"):#set of groupby value of same location :
    bhk_stats={}#used to store the mean ,std,count of house in same set of groupby location :
    for bhk,bhk_df in location_df.groupby("bhk"):
      bhk_stats[bhk]={
          "mean":np.mean(bhk_df.price_per_sqft),
          "std":np.std(bhk_df.price_per_sqft),
          "count":bhk_df.shape[0]
          }
    for bhk,bhk_df in location_df.groupby("bhk"):#compairing the bhk size in neighbourhood:
      stats=bhk_stats.get(bhk-1)
      if stats and stats['count']>5:
        exclude_indices=np.append(exclude_indices,bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
  return df.drop(exclude_indices,axis='index')


  df=bhk_outlier_remove(df)
```

# Split the data into X,y:


```python
X=df.drop(columns=['price'])
y=df['price']
```

```python
#here importing all important libraries of sklearn :
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
#splitting here:
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
print(X_train.shape,X_test.shape,X.shape,y.shape)
```

```python
#here making the columns_transformer :
columns_trans=make_column_transformer((OneHotEncoder(sparse_output=False),['location']),
                                       ( StandardScaler(),['total_sqft']),
                                      remainder='passthrough')
```

```python
#selecting multiple model in one place :
model={
    "Linear Regression":LinearRegression(),
    "Lasso Regression ":Lasso(),
    "Ridge Rgression ":Ridge()
}
results={}

for name,model_instance in model.items():
  pipe=make_pipeline(columns_trans,model_instance)
  pipe.fit(X_train,y_train)
  y_pred=pipe.predict(X_test)
  score=r2_score(y_test,y_pred)
  results[name] = score
  print(f"✅ {name} trained. R2 Score: {score:.4f}")
```

```python
#viewingh the winner here:
summary_df=pd.DataFrame(list(results.items()),columns=['Model','R2_Score'])
summary_df=summary_df.sort_values(by='R2_Score',ascending=False)
print(summary_df.to_string(index=False))
```

# Hyperparameter tunning to increase the R2_score

```python
#perfoming the hyperparameter tunning in Ridge Regression :
from sklearn.model_selection import GridSearchCV
#defining the parameter of ridge regression:
ridge_params={'ridge__alpha':[0.1,1.0,10.0,100.0]}
#pipeline of this :
pipeline_ridge=make_pipeline(columns_trans,Ridge())
#Perform gride_search cv:
gride_ridge=GridSearchCV(pipeline_ridge,ridge_params,cv=5,scoring='r2',n_jobs=-1,verbose=1)
#fit the pipeline with gridesearchcv:
gride_ridge.fit(X_train, y_train)

#selecting the best model and r2_score:
best_ridge=gride_ridge.best_estimator_
best_ridge_r2=gride_ridge.best_score_

#showing in screen:
results['Ridge Rgression (Tunned)']=best_ridge_r2
```

```python
#viewing the score:
summary_df_final_ridge=pd.DataFrame(list(results.items()),columns=['Model','R2_Score'])
summary_df_final_ridge=summary_df_final_ridge.sort_values(by="R2_Score",ascending=False)
print(summary_df_final_ridge.to_string(index=False))
```

```python
#hyperparamter tunning with lasso regression :
#lasso params:
lasso_params={'lasso__alpha':[0.001,0.01,0.1,1.0,10.0]}
#pipeline for lasso :
pipeline_lasso=make_pipeline(columns_trans,Lasso())
#gridsearchcv for lasso :
gride_lasso=GridSearchCV(pipeline_lasso,lasso_params,cv=5,scoring ='r2',n_jobs=-1,verbose=1)
#fitting the model:
gride_lasso.fit(X_train,y_train)
#selecting the best model and r2_score:
best_lasso=gride_lasso.best_estimator_
best_lasso_r2=gride_lasso.best_score_

#shopwing the results:
results['Lasso Regression (Tunned)']=best_lasso_r2


#showing the results :
summary_df_final_lasso=pd.DataFrame(list(results.items()),columns=['Model','R2_Score'])
summary_df_final_lasso=summary_df_final_lasso.sort_values(by="R2_Score",ascending=False)
print(summary_df_final_lasso.to_string(index=False))
```
